```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
def EI(M, base=None):
    M = np.array(M)
    M_length = len(M)
    if base:#是否为2进制
        sums = np.sum([
            0 if M[i, j] == 0 else M[i, j] * np.log(M[i, j] / np.sum(M[:, j], axis=0)) / np.log(base) 
            for i in range(M_length) for j in range(M_length)
        ]) / M_length
        return sums
    else:
        py = np.sum(M, axis=0) / M_length
        pyy = np.tile(py, (M_length, 1))
        ei = np.sum(M * (np.log2(M + 1e-10) - np.log2(pyy + 1e-10)) / M_length)
        return ei

    
def calc_ei(pxy):
    # normalizing p(x, y)
    p = pxy / pxy.sum()
    
    # dimension of x and y
    m, n = p.shape
    
    # marginal distributions
    px = p.sum(axis=1)
    py = p.sum(axis=0)
    
    # do(x) to uniform distribution, tpm = p(y|x)
    tpm = (p.T / px).T
    pu = tpm / tpm.sum()
    
    # marginal distribution of y given x ~ Unifrom Dist
    puy = pu.sum(axis=0)
    
    # replace 0 to a small positive number to avoid log error
    eps = 1E-10
    tpm_e = np.where(tpm==0, eps, tpm)
    puy_e = np.where(tpm==0, eps, puy)
    
    # calculate EI of specific x
    ei_x = (np.log2(tpm_e / puy_e) * tpm).sum(axis=1)
    
    # calculate determinism and degeneracy coefficients of specific x
    det_x = (np.log2(tpm_e * m) * tpm).sum(axis=1) / np.log2(m)
    deg_x = (np.log2(puy_e * m) * tpm).sum(axis=1) / np.log2(m)
    
    # calculate total EI
    ei_all = ei_x.mean()
    
    # calculate determinism and degeneracy coefficients, and efficiency
    det = det_x.mean()
    deg = deg_x.mean()
    eff = det - deg
    
    return tpm, eff, det, deg, ei_all, ei_x, det_x, deg_x


from math import log
def Degeneracy(M, base=2):
    # Ensure M is a numpy array for vectorized operations
    M = np.array(M)
    n = M.shape[0]  # Assuming M is a square matrix
    
    # Compute the column sums
    col_sums = M.sum(axis=0)
    # Calculate the degeneracy measure
    # We use np.where to avoid division by zero or log of zero
    terms = np.where(M != 0, M * np.log(col_sums) / np.log(base), 0)
    sums = np.sum(terms) / (n * log(n, base))
    
    return sums

#plot
def set_plot_info(xlabel,ylabel,title='',fontsize=14):
    plt.xlabel(xlabel,fontsize=fontsize)
    plt.ylabel(ylabel,fontsize=fontsize)
    plt.title(title,fontsize=fontsize)
    plt.xticks(fontsize=fontsize-2)
    plt.yticks(fontsize=fontsize-2)
```

```python
# Perturb Permutation Matrix

# Main computation
xys = []
for sz in [200,100,50,10,2]:
    print(sz)
    xy = []
    for sigma in np.arange(-3, 3.1, 0.1):
        sigma_val = 10 ** sigma
        perms = np.transpose([np.arange(sz), np.random.permutation(sz)])
        A = np.zeros((sz, sz))
        for i in range(len(perms)):
            ys = stats.norm.pdf(np.arange(sz), loc=perms[i, 1], scale=sigma_val*sz/6)
            A[i, :] = ys
        M = A / np.sum(A, axis=1)[:, np.newaxis]
        singular = np.linalg.svd(M, compute_uv=False)
#         xx_0 = np.sum(singular**0)
        xx_1 = np.sum(singular**0.1)
        xx_2 = np.sum(singular**1)
#         xx_3 = np.sum(singular**1.5)
        xx_4 = np.sum(singular**1.9)
        ei = EI(M)
        xys.append([ei, sigma_val, xx_1, xx_2, xx_4, sz])
#     xys.append([sz, xy])


xys = np.array(xys)
xys.shape
```

## Controlled Degeneracy

```python
from scipy.linalg import svd
from scipy.stats import norm

# Parameters
alpha = 1
sz = 50
errors = []

# Generate the data
xys = []
for sigma1 in np.arange(-3, 3.1, 0.1):
    sigma = 10**sigma1
    print(sigma)
    sigma_results = []
    for bits in range(0, sz//2 + 1):
        perms = np.column_stack((np.arange(sz), np.concatenate((np.ones(bits), np.arange(sz - bits) + bits))))
        A = np.zeros((sz, sz))

        for i in range(sz):
            ys = norm.pdf(np.arange(sz), loc=perms[i, 1], scale=sigma * sz / 6)
            A[i, :] = ys

        M = A / np.sum(A, axis=1, keepdims=True)
        singular_values = svd(M, compute_uv=False)
        xx = np.sum(singular_values**alpha)
        yy = EI(M)
        deg = Degeneracy(M)

        if yy > np.log2(xx) or yy < np.log2(xx) - np.log2(sz):
            print(f"OMG! {sigma}, Deg={deg}, logΓ={np.log2(xx)}, EI={yy}, logN={np.log2(sz)}")
            errors.append(M)
        sigma_results.append([xx, yy, deg, bits, sigma])

    xys.extend(sigma_results)
xys = np.array(xys)
# Flatten the results if needed or process them as they are
```

```python
counts = 6
plt.figure(dpi=100)
markers = ['o','s','^','*','p']
colors = [plt.cm.Blues_r(i / counts) for i in range(counts)]
for idx,r in enumerate([0,5,10,15,25]): #选取rank
    xy = xys[xys[:,3] == r]
    plt.plot(xy[:,0], xy[:,1],color=colors[idx],label=f'r={50-r}',marker=markers[idx],markersize=7)
x = range(int(xys.min()),int(xys.max()))
uplimit = np.log2(x)*2/alpha
plt.plot(x,uplimit,':',color='red')
plt.plot(x, np.log2(x), ':', color='grey')
plt.plot(x, np.log2(x)-np.log2(sz), ':',color='red')
plt.text(8,5.4,r'$2 \log(\Gamma)$',fontsize=14)
plt.text(15,4.5,r'$\log(\Gamma)$',fontsize=14)
plt.text(15,0,r'$\log(\Gamma) - \logN$',fontsize=14)
plt.legend(fontsize=14)
plt.ylim(-2,6.2)
plt.xlim(0,55)
set_plot_info(r'Dynamics Reversibility $\Gamma$','Effective Information (EI)',fontsize=20)
plt.show()
```

```python

def compute_xys(sz_range, m_iterations):
    xys = []
    '''
    sz_range: size range of matrix,
    m_iterations: num. of random samples
    
    return:
    dim0=sum of singular;dim1=eff
    '''
    alpha = 1
    xys = np.zeros([sz_range[1]+1-sz_range[0],m_iterations, 2])
#     print(xys.shape)
#     de
    count = 0
    for sz in tqdm(range(sz_range[0], sz_range[1] + 1)):
        for m in range(m_iterations):
        
            A = np.random.rand(sz, sz)
            M = A / np.sum(A, axis=1)[:, np.newaxis] #归一化矩阵使之成为转移矩阵
            eigen = np.linalg.svd(M, compute_uv=False)
            xx = np.sum(eigen**alpha)
            yy = EI(M)
            xys[count,m,0] = xx
            xys[count,m,1] = yy
        count += 1
    return xys

# Computing xys
sz_range = [2,100]
xys = compute_xys(sz_range, 100)
```
