```python
def calc_ei(pxy):
    # normalizing p(x, y)
    p = pxy / pxy.sum()
    
    # dimension of x and y
    m, n = p.shape
    
    # marginal distributions
    px = p.sum(axis=1)
    py = p.sum(axis=0)
    
    # do(x) to uniform distribution, tpm = p(y|x)
    tpm = (p.T / px).T
    pu = tpm / tpm.sum()
    
    # marginal distribution of y given x ~ Unifrom Dist
    puy = pu.sum(axis=0)
    
    # replace 0 to a small positive number to avoid log error
    eps = 1E-10
    tpm_e = np.where(tpm==0, eps, tpm)
    puy_e = np.where(tpm==0, eps, puy)
    
    # calculate EI of specific x
    ei_x = (np.log2(tpm_e / puy_e) * tpm).sum(axis=1)
    
    # calculate determinism and degeneracy coefficients of specific x
    det_x = (np.log2(tpm_e * m) * tpm).sum(axis=1) / np.log2(m)
    deg_x = (np.log2(puy_e * m) * tpm).sum(axis=1) / np.log2(m)
    
    # calculate total EI
    ei_all = ei_x.mean()
    
    # calculate determinism and degeneracy coefficients, and efficiency
    det = det_x.mean()
    deg = deg_x.mean()
    eff = det - deg
    
    return tpm, eff, det, deg, ei_all, ei_x, det_x, deg_x
```

```python

def pertub_tpm(tpm, pertubation=0):
    """
    Pertub a TPM.
    @param tpm: a TPM
    @param pertubation: the pertubation strength (standard deviation of the normal distribution)
    @return: a pertubed TPM
    """

    A = tpm.copy()
    size = A.shape[0]
    for i in tqdm(range(size)):
        while True:
            indexs = np.where(A[i,:] > 0)[0]
            pertub = np.random.normal(size=(indexs.shape)) * pertubation
            pertub = pertub - pertub.mean()
            if ((A[i,indexs] + pertub) > 0).all():
                break
        
        A[i,indexs] += pertub
    return A
```

```python
# construct diagnal block matrix

def diag_block(matrices):
    """
    Construct a diagnal block matrix.
    @param matrices: a list of matrices
    @return: a diagnal block matrix
    """
    
    comb = []
    for i in range(len(matrices)):
        row = []
        for j in range(len(matrices)):
            if i == j:
                row.append(matrices[i])
            else:
                row.append(np.zeros((matrices[i].shape[0], matrices[j].shape[1])))
        comb.append(row)
    return np.block(comb)


def permutation(matrix, row_only=False):
    """
    Permutate the order of a given matrix.
    @param matrix: matrix to be permutated
    @param row_only: only permutate the rows if True, otherwise permutate both rows and columns
    @return: a permutated matrix
    """
    
    idx = np.random.permutation(matrix.shape[0])
    if row_only:
        return matrix[idx, :]
    else:
        return matrix[idx, :][:, idx]
```

# Clustering based on the distence of row vector distance

```python
def grouping_rows(a, tolerance=0.01):
    """
    a: tmp,
    tolerance: threshold
    Group rows of a matrix.
    """
    # calculate the distance matrix
    size = a.shape[0]
    # convert row vectors to unit vectors
    a_norm = a ** 0.5

    # calculate the distance matrix
    a_dist = a_norm @ a_norm.T
    # mapping the distance matrix to a list of indices tuples
    distances = [{'pos':(i,j), 'dist':a_dist[i,j]} for i,j in itertools.combinations(range(size), 2)]
    distances += [{'pos':(i,i), 'dist':a_dist[i,i]} for i in range(size)]
    
    # group similar rows
    similar_pairs = [d['pos'] for d in distances if 1 - d['dist'] < tolerance]
    
    groups = []
    'grouping'
    for pair in similar_pairs:
        registered = False
        for g in groups:
            if pair[0] in g or pair[1] in g:
                g.add(pair[0])
                g.add(pair[1])
                registered = True
                break
        if not registered:
            groups.append(set(pair))
#     print(groups)
    for i in range(len(groups)):
        for j in range(i+1, len(groups)):
            if groups[i] & groups[j]:
                groups[i] |= groups[j]
                groups[j] = set()
                
    groups = [g for g in groups if g]
    permut = [x for g in groups for x in g] 
    permuted_a = a[permut, :][:, permut]
    permuted_norm = a_dist[permut, :][:, permut]
    
    group_matrix = np.zeros([size,len(groups)])
    

    for i in range(len(groups)):
        for j in groups[i]:
            group_matrix[j,i] = 1

    return groups, permuted_a, group_matrix#permuted_norm
```

# grouping matrix

```python
def group_transform(groups):
    """
    Construct a transformation matrix for a given group.
    """
    size = sum([len(g) for g in groups])
    transform = np.zeros((size, len(groups)))
    for i, g in enumerate(groups):
        for j in g:
            transform[j, i] = 1
    return transform
```

```python
def plot_hm(mi_tpm, annot=True, font_scale=1, colorbar_formatter=lambda x, pos:f"{x:.2f}"):
    if mi_tpm.shape[0] > 10:
        annot = False
    sns.heatmap(mi_tpm, annot=annot, cmap='Blues')

def plot_group(a, tolerance=0.01):
    a_grp, perm_a, group_matrix = grouping_rows(a, tolerance)
#     print('Grouping', a_grp)
    plt.figure(dpi=100,figsize=(16, 4))
    plt.subplot(1, 3, 1)
    plot_hm(a)
    
    plt.title(f'Original TPM,EI:{calc_ei(a)[4]:.4f}')
    plt.subplot(1, 3, 2)
    plot_hm(group_matrix)
    plt.title('Grouping matrix')
    coarse_grain_matrix = ((group_matrix.T)@a@group_matrix) * 1/group_matrix.sum(axis=0)
    plt.subplot(1, 3, 3)
    plot_hm(coarse_grain_matrix)
    plt.title(f'Coarse grain TPM,EI:{calc_ei(coarse_grain_matrix)[4]:.4f}')
```

```python
def generate_transition_matrix():
    """
    Generate a 64x64 transition matrix for the Boolean network based on the defined rules.
    
    Returns:
    numpy.ndarray: A 64x64 transition matrix.
    """
    # Define the probabilities for spontaneous activation
    p_A, p_B, p_C = 0.45, 0.5, 0.55
    p_D, p_E, p_F = p_A, p_B, p_C  # Same probabilities for group DEF
    
    # Initialize the transition matrix with zeros
    transition_matrix = np.zeros((64, 64))
    
    # Loop over all possible states
    for current_state_index in range(64):
        # Convert the current state index to a binary string
        current_state_bin = format(current_state_index, '06b')
        
        # Calculate the sum of intragroup connections for both groups
        sum_intra_ABC = sum(int(current_state_bin[i]) for i in range(3))
        sum_intra_DEF = sum(int(current_state_bin[i+3]) for i in range(3))
        
        # Check the conditions for each group based on the sum of intragroup connections
        # and apply the intergroup connection rules
        for next_state_index in range(64):
            # Convert the next state index to a binary string
            next_state_bin = format(next_state_index, '06b')
            
            # Start with a probability of 1 and adjust based on the rules
            prob = 1.0
            
            # Check group ABC
            if sum_intra_ABC == 0:
                # If Σ(intra) for ABC is 0, then A, B, C stay 0
                if any(int(next_state_bin[i]) for i in range(3)):
                    prob = 0  # Any of A, B, C being 1 is not possible
            else:
                # Apply spontaneous activation probabilities for A, B, C
                for i in range(3):
                    if current_state_bin[i] == '0':
                        prob *= p_A if next_state_bin[i] == '1' else (1 - p_A)
            
            # Check group DEF
            if sum_intra_DEF == 0:
                # If Σ(intra) for DEF is 0, then D, E, F stay 0
                if any(int(next_state_bin[i+3]) for i in range(3)):
                    prob = 0  # Any of D, E, F being 1 is not possible
            else:
                # Apply spontaneous activation probabilities for D, E, F
                for i in range(3, 6):
                    if current_state_bin[i] == '0':
                        prob *= p_D if next_state_bin[i] == '1' else (1 - p_D)
            
            # Assign the calculated probability to the transition matrix
            transition_matrix[current_state_index, next_state_index] = prob
    
    return transition_matrix

# Generate the transition matrix
transition_matrix = generate_transition_matrix()

# Example usage:
# Get the transition probabilities for the current state '000001'
current_state = '000001'
current_state_index = int(current_state, 2)
transition_probabilities_for_current_state = transition_matrix[current_state_index]

transition_probabilities_for_current_state
```

```python
# Define the Transition function according to the provided logic
def Transition(state1, state2, nd):
    if sum(state2) == 3:
        results = [0, 1]
    elif sum(state1) == 0:
        results = [1, 0]
    else:
        if nd in [1, 4, 7]:
            results = [0.55, 0.45]
        elif nd in [2, 5, 8]:
            results = [0.5, 0.5]
        elif nd in [3, 6, 9]:
            results = [0.45, 0.55]
    return results

# Implement the OneStep function
def OneStep(states):
    neighbors1 = [[1, 2, 3], [1, 2, 3], [1, 2, 3], [4, 5, 6], [4, 5, 6], [4, 5, 6]]
    neighbors2 = [[4, 5, 6], [4, 5, 6], [4, 5, 6], [1, 2, 3], [1, 2, 3], [1, 2, 3]]
    result = []
    for i, _ in enumerate(states):
        state1 = [states[j-1] for j in neighbors1[i]] # Adjust for Python's 0-indexing
        state2 = [states[j-1] for j in neighbors2[i]] # Adjust for Python's 0-indexing
        transition_result = Transition(state1, state2, i+1) # Adjust for Python's 0-indexing
        result.append(transition_result)
    return np.array(result)

# Rewrite the Pxy function to use the OneStep function
def Pxy(allstates):
    results = []
    for i, state_i in enumerate(allstates):
        probs = OneStep(state_i)
        state_results = []
        for j, state_j in enumerate(allstates):
            digits = np.ravel(state_j)
            temp = np.prod([probs[m, digit] for m, digit in enumerate(digits)])
            state_results.append(temp)
        results.append(state_results)
    return results

# Example usage
allstates = np.array([list(map(int, format(i, '06b'))) for i in range(2**6)])
# print(allstates)
  # This is just a placeholder. Replace it with your actual list of states.
pxy = np.array(Pxy(allstates))
plt.figure(dpi=100)
sns.heatmap(pxy, cmap='Blues')
plt.title(f'EI:{calc_ei(pxy)[4]:.4f}',fontsize=14)
plt.xticks([])
plt.yticks([])
plt.show()
```
