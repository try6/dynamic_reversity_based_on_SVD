```python
import numpy as np
import pandas as pd
import itertools
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
import matplotlib as mpl
from tqdm import tqdm
```

```python
def set_plot_info(xlabel,ylabel,fontsize=14):
    plt.xlabel(xlabel,fontsize=fontsize)
    plt.ylabel(ylabel,fontsize=fontsize)
    plt.xticks(fontsize=fontsize)
    plt.yticks(fontsize=fontsize)
    
def calc_ei(pxy):
    # normalizing p(x, y)
    p = pxy / pxy.sum()
    
    # dimension of x and y
    m, n = p.shape
    
    # marginal distributions
    px = p.sum(axis=1)
    py = p.sum(axis=0)
    
    # do(x) to uniform distribution, tpm = p(y|x)
    tpm = (p.T / px).T
    pu = tpm / tpm.sum()
    
    # marginal distribution of y given x ~ Unifrom Dist
    puy = pu.sum(axis=0)
    
    # replace 0 to a small positive number to avoid log error
    eps = 1E-10
    tpm_e = np.where(tpm==0, eps, tpm)
    puy_e = np.where(tpm==0, eps, puy)
    
    # calculate EI of specific x
    ei_x = (np.log2(tpm_e / puy_e) * tpm).sum(axis=1)
    
    # calculate determinism and degeneracy coefficients of specific x
    det_x = (np.log2(tpm_e * m) * tpm).sum(axis=1) / np.log2(m)
    deg_x = (np.log2(puy_e * m) * tpm).sum(axis=1) / np.log2(m)
    
    # calculate total EI
    ei_all = ei_x.mean()
    
    # calculate determinism and degeneracy coefficients, and efficiency
    det = det_x.mean()
    deg = deg_x.mean()
    eff = det - deg
    
    return tpm, eff, det, deg, ei_all, ei_x, det_x, deg_x


```

```python
# pertub a TPM

def pertub_tpm(tpm, pertubation=0):
    """
    Pertub a TPM.
    @param tpm: a TPM
    @param pertubation: the pertubation strength (standard deviation of the normal distribution)
    @return: a pertubed TPM
    """

    A = tpm.copy()
    size = A.shape[0]
    for i in tqdm(range(size)):
#         print(i)
        while True:
            indexs = np.where(A[i,:] > 0)[0]
#             print(indexs,indexs.shape)
#             de
            pertub = np.random.normal(size=(indexs.shape)) * pertubation
            pertub = pertub - pertub.mean()
            if ((A[i,indexs] + pertub) > 0).all():
                break
        
        A[i,indexs] += pertub
    return A
```

```python
# construct a random TPM with a given number of causal states
def random_states_tpm(size, rank=None):
    """
    Generate a random TPM with a given number of causal states.
    @param size: the size of the TPM
    @param rank: the number of causal states
    @return: a random TPM with a given number of causal states
    """
    
    rank = rank if rank else size
    # 创建一个n*r的随机矩阵A，其中每一行的元素之和为1
    A = np.random.normal(size=(rank, size))
    A /= np.linalg.norm(A, axis=1, keepdims=True)
    A **= 2
 
    if rank < size:
        fill = np.random.choice(rank, size-rank)
        A = np.concatenate([A, A[fill, :]], axis=0)
        A = np.random.permutation(A)
        
    return A
```

```python
def get_EI(M, base=None):
    M = np.array(M)
    M_length = len(M)
    if base:#是否为2进制
        sums = np.sum([
            0 if M[i, j] == 0 else M[i, j] * np.log(M[i, j] / np.sum(M[:, j], axis=0)) / np.log(base) 
            for i in range(M_length) for j in range(M_length)
        ]) / M_length
        return sums
    else:
        py = np.sum(M, axis=0) / M_length
        pyy = np.tile(py, (M_length, 1))
        ei = np.sum(M * (np.log2(M + 1e-10) - np.log2(pyy + 1e-10)) / M_length)
        return ei
```
